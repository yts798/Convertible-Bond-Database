```python
# set up
import pandas as pd
import numpy as np
import pickle
import time 
import os
from datetime import datetime
from operator import itemgetter
data_folder_path = 'C:\\Users\\goodluck\\Desktop\\database_workspace'
raw_data_path =  "C:\\Users\\goodluck\\Desktop\\database_workspace\\raw_data"
storage_path =  "C:\\Users\\goodluck\\Desktop\\database_workspace\\database_storage"
DTBS_path = os.path.join(storage_path, "DTBS.pkl")
```

```python
# helper functions
def select_wkds(days, name):
    if name == 'mon':
        sel = 0
    elif name == 'tue':
        sel = 1
    elif name == 'wed':
        sel = 2
    elif name == 'thu':
        sel = 3
    elif name == 'fri':
        sel = 4
    
    dts = []
    for i in days:
        dts.append(datetime.strptime(i, '%Y-%m-%d').date())
        
    sel_dts = [d for d in dts if d.weekday() == sel]

    selected_dates = [i.strftime('%Y-%m-%d') for i in sel_dts]
    
    return selected_dates



def db_low(obj):
    cpr = obj['cpr']
    dp = obj['dp']
    if np.isnan(cpr):
        return 999

    elif np.isnan(dp):
        return 999
    else:
        return cpr + dp
    
def takeSecond(elem):
    return elem[1]
```

```python
# read raw data
cpr_path =  os.path.join(raw_data_path, "cpr.CSV")
dp_path = os.path.join(raw_data_path, "dp.CSV")
zi_path = os.path.join(raw_data_path, "zi.CSV")
binfo_path = os.path.join(raw_data_path, "binfo.CSV")
bl_path = os.path.join(raw_data_path, "bl.CSV")
trt_path = os.path.join(raw_data_path, "trt.CSV")
yl_path = os.path.join(raw_data_path, "yl.CSV")
ytm_path = os.path.join(raw_data_path, "ytm.CSV")
cpr_df = pd.read_csv(cpr_path)
dp_df = pd.read_csv(dp_path)
zi_df = pd.read_csv(zi_path)
binfo_df = pd.read_csv(binfo_path, engine='python')
bl_df = pd.read_csv(bl_path)
trt_df = pd.read_csv(trt_path)
yl_df = pd.read_csv(yl_path)
ytm_df = pd.read_csv(ytm_path)
cpr_df.set_index('Date',inplace=True)
dp_df.set_index('Date',inplace=True)
zi_df.set_index('Date',inplace=True)
binfo_df.set_index('Date',inplace=True)
bl_df.set_index('Date',inplace=True)
trt_df.set_index('Date',inplace=True)
ytm_df.set_index('Date',inplace=True)
yl_df.set_index('Date',inplace=True)
```

```python
# initialise structure
DTBS = dict()
DTBS['A'] = dict()
DTBS['B'] = dict()
DTBS['C'] = dict()
DTBS['D'] = dict()
```

```python
# build A, B structure
for code in cpr_df.columns:
    DTBS['A'][code] = dict()
    DTBS['B'][code] = dict()
    for date in cpr_df.index:
        DTBS['A'][code][date] = dict()
        
        
```

```python
# save cpr, dp (for A)
count = 0
for code in cpr_df.columns:

    for date in cpr_df.index:
        DTBS['A'][code][date] = dict()
        DTBS['A'][code][date]['cpr'] = cpr_df.loc[date][code]
        DTBS['A'][code][date]['dp'] = dp_df.loc[date][code]

        DTBS['A'][code][date]['bl'] = bl_df.loc[date][code]
        DTBS['A'][code][date]['trt'] = trt_df.loc[date][code]   
        DTBS['A'][code][date]['ytm'] = ytm_df.loc[date][code]
        DTBS['A'][code][date]['yl'] = yl_df.loc[date][code]
    count += 1
    print(count)
```

```python
# save zi (for C)

DTBS['C']['zi'] = dict()
for date in zi_df.index:
    DTBS['C']['zi'][date] = zi_df.loc[date]['000832.CSI']
    
```

```python
# save dates, each monday to friday (for D)
DTBS['D']['day'] = list(cpr_df.index)
DTBS['D']['mon'] = select_wkds(list(cpr_df.index), 'mon')
DTBS['D']['tue'] = select_wkds(list(cpr_df.index), 'tue')
DTBS['D']['wed'] = select_wkds(list(cpr_df.index), 'wed')
DTBS['D']['thu'] = select_wkds(list(cpr_df.index), 'thu')
DTBS['D']['fri'] = select_wkds(list(cpr_df.index), 'fri')
```

```python
# clear noisy data before ipo
# clean wrong 100 dirtyprice before IPO date
for code in cpr_df.columns:

    ipodt = DTBS['B'][code]['ipo']
    dt = datetime.strptime(ipodt, '%Y-%m-%d').date()
        
    dtss = list(DTBS['A'][code].keys())
    f_dttm = datetime.strptime(dtss[0], '%Y-%m-%d').date()
    if f_dttm >= dt:
        continue
    for tm in dtss:
        
        dttm = datetime.strptime(tm, '%Y-%m-%d').date()
        if dttm == dt:
            break
        
        DTBS['A'][code][tm]['cpr'] = np.nan
        DTBS['A'][code][tm]['dp'] = np.nan     
        DTBS['A'][code][tm]['yl'] = np.nan
        

        

```

```python
# calculate and save double low
for code in cpr_df.columns:
    for date in cpr_df.index:
        DTBS['A'][code][date]['dl'] = db_low(DTBS['A'][code][date])
```

```python
# save
f_save = open(DTBS_path, 'wb')
pickle.dump(DTBS, f_save)
f_save.close()

```
