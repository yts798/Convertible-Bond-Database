```python
cz_path =  os.path.join(auto_path, "cz06-27-07-03.csv")
cz_df = pd.read_csv(cz_path, index_col = 0)
cz_df
```

```python
count = 0
for code in DTBS['A'].keys():
    count += 1
    for date in DTBS['D']['day']:
        if DTBS['A'][code][date]['ia'] == 1:
            DTBS['A'][code][date]['csv'] = ((100.0 / DTBS['A'][code][date]['csp']) *   DTBS['E'][DTBS['B'][code]['sc']][date]['cl'])
            
        else:
            DTBS['A'][code][date]['csv'] = 0
    print(count)
```

```python
for code in cz_df.columns:
    count = 0
    for date in dts:
        if DTBS['A'][code][date]['ia'] == 1:
            DTBS['A'][code][date]['cz'] = cz_df.iloc[count][code]
            
        else:
            DTBS['A'][code][date]['cz'] = 0
        count += 1
```

```python
for code in cz_df.columns:
    for date in dts:
        if DTBS['A'][code][date]['ia'] == 1:
            if code == '110057.SH' or code == '113016.SH':
                DTBS['A'][code][date]['pd'] = 0
                
                # 这里是一个坑，在cz wind api里面应该每天一行一行的添加，而不是按五天标准，不然强赎会有问题
            else:
#                 print(code)
                DTBS['A'][code][date]['pd'] = ((DTBS['A'][code][date]['csv'] /  DTBS['A'][code][date]['cz'] ) - 1.0) * 100
            
        else:
            DTBS['A'][code][date]['pd'] = 0
```
