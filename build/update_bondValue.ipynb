```python
# set up
import pandas as pd
import numpy as np
import pickle
import time 
import os
from datetime import datetime
from operator import itemgetter
data_folder_path = 'C:\\Users\\goodluck\\Desktop\\DB'
raw_data_path =  "C:\\Users\\goodluck\\Desktop\\DB\\raw_data"
storage_path =  "C:\\Users\\goodluck\\Desktop\\DB\\database_storage"
DTBS_path = os.path.join(storage_path, "DTBS.pkl")

with open(DTBS_path, 'rb') as f:  
    DTBS = pickle.load(f)
    
    
```

```python
cz_path =  os.path.join(raw_data_path, "cz.CSV")
cz_df = pd.read_csv(cz_path)
cz_df.set_index('Date',inplace=True)
cz_df
```

```python
count = 0
for code in DTBS['A'].keys():
    count += 1
    for date in DTBS['D']['day']:
        if DTBS['A'][code][date]['ia'] == 1:
            DTBS['A'][code][date]['csv'] = ((100.0 / DTBS['A'][code][date]['csp']) *   DTBS['E'][DTBS['B'][code]['sc']][date]['cl'])
            
        else:
            DTBS['A'][code][date]['csv'] = 0
    print(count)
```

```python
dts = DTBS['D']['day'][:1537]
dts[-1]
# must be before 2023/4/28
```

```python
for code in cz_df.columns:
    count = 0
    for date in dts:
        if DTBS['A'][code][date]['ia'] == 1:
            DTBS['A'][code][date]['cz'] = cz_df.iloc[count][code]
            
        else:
            DTBS['A'][code][date]['cz'] = 0
        count += 1
```

```python
for code in cz_df.columns:
    for date in dts:
        if DTBS['A'][code][date]['ia'] == 1:
            DTBS['A'][code][date]['pd'] = ((DTBS['A'][code][date]['csv'] /  DTBS['A'][code][date]['cz'] ) - 1.0)
            
        else:
            DTBS['A'][code][date]['pd'] = 0
```
