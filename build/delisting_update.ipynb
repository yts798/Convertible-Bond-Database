```python
# before and include 3.10
count = 0
for code in cpr_df.columns:
    DTBS['A'][code] = dict()
    for dt in cpr_df.index:
        date = datetime.strptime(dt, '%Y/%m/%d').strftime('%Y-%m-%d')
        DTBS['A'][code][date] = dict()
        DTBS['A'][code][date]['cpr'] = cpr_df.loc[dt][code]
        DTBS['A'][code][date]['dp'] = dp_df.loc[dt][code]

        DTBS['A'][code][date]['bl'] = bl_df.loc[dt][code]
        DTBS['A'][code][date]['trt'] = trt_df.loc[dt][code]   
        DTBS['A'][code][date]['ytm'] = ytm_df.loc[dt][code]
        
        DTBS['A'][code][date]['yl'] = yl_df.loc[dt][code]
    count += 1
    print(count)
```

```python
count = 0
dtss = DTBS['D']['day']
for code in list(DTBS['B'].keys()):
    ipo = DTBS['B'][code]['ipo']
    dld = DTBS['B'][code]['dld']
    ipodt = datetime.strptime(ipo, '%Y-%m-%d').date()
    dlddt = datetime.strptime(dld, '%Y-%m-%d').date()
        
    for tm in dtss:
        
        dttm = datetime.strptime(tm, '%Y-%m-%d').date()
        if dttm < ipodt:
            DTBS['A'][code][tm]['cpr'] = np.nan
            DTBS['A'][code][tm]['dp'] = np.nan     
            DTBS['A'][code][tm]['yl'] = np.nan
            DTBS['A'][code][tm]['ytm'] = np.nan
            DTBS['A'][code][tm]['bl'] = np.nan
            DTBS['A'][code][tm]['trt'] = np.nan
            DTBS['A'][code][tm]['ia'] = 0
            
        elif dttm > dlddt:
            DTBS['A'][code][tm]['cpr'] = np.nan
            DTBS['A'][code][tm]['dp'] = np.nan     
            DTBS['A'][code][tm]['yl'] = np.nan
            DTBS['A'][code][tm]['ytm'] = np.nan
            DTBS['A'][code][tm]['bl'] = np.nan
            DTBS['A'][code][tm]['trt'] = np.nan
            DTBS['A'][code][tm]['ia'] = 0
        else:
            DTBS['A'][code][tm]['ia'] = 1
            
    count += 1
    print(count)
            
            
        

        
        
```

```python

def db_low(obj):
    cpr = obj['cpr']
    dp = obj['dp']
    if np.isnan(cpr):
        return 999

    elif np.isnan(dp):
        return 999
    else:
        return cpr + dp
    
for code in list(DTBS['B'].keys()):
    for date in DTBS['D']['day']:
        DTBS['A'][code][date]['dl'] = db_low(DTBS['A'][code][date])
```
