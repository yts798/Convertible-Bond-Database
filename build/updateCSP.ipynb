```python
csp_path =  os.path.join(raw_data_path, "csp.CSV")
csp_df = pd.read_csv(csp_path)
csp_df.set_index('Date',inplace=True)
csp_df
```

```python
for code in csp_df.columns:
    for date in csp_df.index:
        if DTBS['A'][code][date]['ia'] == 1:
            DTBS['A'][code][date]['csp'] = csp_df.loc[date][code]
            
        else:
            DTBS['A'][code][date]['csp'] = 0
```
