```python
import pandas as pd
import numpy as np
import pickle
import time 
import os
import re
from datetime import datetime
```

```python
# load
sdgd_df = pd.read_csv('十大持股结果20230612.CSV', encoding="gbk")
sdgd_df.set_index('2023/6/12',inplace=True)
nmlst = ['n'+str(i) for i in range(1,11)]
nmlst

sdgd_df
```

```python
# clean empty
name_pool = []
for code in sdgd_df.columns:
    for nm in nmlst:
        rn = 'r' + nm[1:]
        if sdgd_df[code][rn] == '0.0000':
#             print(sdgd_df[code][nm])
            sdgd_df[code][nm] = '0'
            
```

```python
new_data = []
for code in sdgd_df.columns:
    new_dgd = []
    for nm in nmlst:
        rn = 'r' + nm[1:]

        if ';' in sdgd_df[code][nm]:
            dgds = sdgd_df[code][nm].split(';')
            for one_ele in dgds:
                new_dgd.append(sdgd_df[code][rn])
                new_dgd.append(one_ele)
        
        else:
            new_dgd.append(sdgd_df[code][rn])
            new_dgd.append(sdgd_df[code][nm])
            
            
    new_20 = new_dgd[:20]
    new_data.append(new_20)
    
print(len(new_data))
    
    
    
```

```python
npnew = np.array(new_data)
new_df = pd.DataFrame(npnew.T,index=sdgd_df.index,columns=sdgd_df.columns)
new_df.to_csv('cleaned.CSV', encoding="gbk")
```
